```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_data = pd.read_csv('data/params_final.csv', sep=';')
lovers = all_data[all_data['avg_rating']>=4]
haters = all_data[all_data['avg_rating']<=2]
others = all_data[(all_data['avg_rating']>2) & (all_data['avg_rating']<4)]
print(all_data.columns)
# lovers =  pd.read_csv('data/brandlovers.csv', sep=';')
# haters = pd.read_csv('data/brandhaters.csv', sep=';')
lovers['avg_sentiment'] = 10*lovers['avg_sentiment']
haters['avg_sentiment'] = 10*haters['avg_sentiment']
```

# Average Upvotes CDF

```python
plt.hist(lovers['avg_upvotes'], density=1, cumulative=True, histtype='step')
plt.hist(haters['avg_upvotes'], density=1, cumulative=True, histtype='step')
```

```python
sns.kdeplot(lovers['avg_upvotes'], cumulative=True)
sns.kdeplot(haters['avg_upvotes'], cumulative=True)
```

```python
a0 = plt.hist(lovers['avg_sentiment'], density=1, cumulative=True, histtype='step', bins=1000)
a2 = plt.hist(haters['avg_sentiment'], density=1, cumulative=True, histtype='step', bins=1000)
```
