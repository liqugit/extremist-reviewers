```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_data = pd.read_csv('data/params.csv', sep=';')
print(all_data.columns)
lovers = all_data[all_data['avg_rating']>=4]
haters = all_data[all_data['avg_rating']<=2]
others = all_data[(all_data['avg_rating']>2) & (all_data['avg_rating']<4)]
lovers['avg_sentiment'] = 10*lovers['avg_sentiment']
haters['avg_sentiment'] = 10*haters['avg_sentiment']
```

# Average Upvotes CDF

```python
plt.hist(lovers['avg_upvotes'], density=1, cumulative=True, histtype='step')
plt.hist(haters['avg_upvotes'], density=1, cumulative=True, histtype='step')
```

```python
sns.kdeplot(lovers['avg_upvotes'], cumulative=True)
sns.kdeplot(haters['avg_upvotes'], cumulative=True)
```

# Avg Sentiment CDF

```python
a0 = plt.hist(lovers['avg_sentiment'], density=1, cumulative=True, histtype='step', bins=1000)
a2 = plt.hist(haters['avg_sentiment'], density=1, cumulative=True, histtype='step', bins=1000)
```

# Avg Review Length

```python
sns.kdeplot(lovers['avg_rev_len'], cumulative=True)
sns.kdeplot(haters['avg_rev_len'], cumulative=True)
```

# Avg Title Length

```python
sns.kdeplot(lovers['avg_title_len'], cumulative=True)
sns.kdeplot(haters['avg_title_len'], cumulative=True)
```


# Avg Title Sentiment

```python
a0 = plt.hist(lovers['avg_title_sentiment'], cumulative=True, density=True, histtype='step', bins=1000)
a2 = plt.hist(haters['avg_title_sentiment'], density=True, cumulative=True, histtype='step', bins=1000)
```

```python
haters['avg_title_len'].hist(density=True, color='r', alpha=0.5, bins=10)
lovers['avg_title_len'].hist(density=True, color='g', alpha=0.5, bins=10)
```
